```python
import numpy as np

# Define Pauli matrices
I = np.array([[1, 0], [0, 1]])
Z = np.array([[1, 0], [0, -1]])
X = np.array([[0, 1], [1, 0]])
Y = np.array([[0, -1j], [1j, 0]])

# Hadamard matrix H
H = (1 / np.sqrt(2)) * (X + Z)

# a) Verify X^2 = Y^2 = Z^2 = I
print("X^2:\n", np.dot(X, X))
print("Y^2:\n", np.dot(Y, Y))
print("Z^2:\n", np.dot(Z, Z))

# b) Verify -iXYZ = I
XYZ = np.dot(X, np.dot(Y, Z))
print("-iXYZ:\n", -1j * XYZ)
```

The rows of the matrix correspond to the enumberated qbits values $|x_1 x_2\rangle$, $|01\rangle$, $|10\rangle$, $|11\rangle$, they are constructed via $x_1 \otimes x_2$
