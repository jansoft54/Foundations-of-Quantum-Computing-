```python
import numpy as np
import cmath
# define the beam splitter matrix
B = (1/np.sqrt(2))*np.array([[1,1j],[1j,1]])

# compute the matrix powers

for i in range(0, 9):
    print("B^"+ str(i) +" = "+ str(np.linalg.matrix_power(B, i)))

print("B^16 = " + str(np.linalg.matrix_power(B, 16)))
print("B^32 = " + str(np.linalg.matrix_power(B, 32)))
```

### Observations
The matrix \( B \) exhibits a periodic behavior with a cycle of 8. This was observed for the first 4 occurrences (8, 16, 24, 36)
Odd powers show alternating signs, either only in the imaginary unit (7) or in the real numbers (3) or in both (5).
The scalar $\frac{1}{\sqrt{2}}$ only appears in odd powers.

```python
import pandas as pd

def vec(x):
    return np.array([1,x])


M = np.array([[0,1],[1,-1]])

x_s = [0, 0, 1, 1]
y_s = [0, 1, 0, 1]

res = []
for i in range(0, 4):
    res.append(np.matmul(vec(x_s[i]).T, np.matmul(M, vec(y_s[i]))))

print(pd.DataFrame({'x': x_s, 'y': y_s, 'x^TMx': res}))
```
