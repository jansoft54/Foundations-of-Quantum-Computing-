We are given V.
 $$ V =
   \frac{1}{\sqrt{2}}
   \begin{bmatrix}
   +i & -i \\
   +1 & +1
   \end{bmatrix}
$$

V is unitary because:

 $$ V^\dag =
   \frac{1}{\sqrt{2}}
   \begin{bmatrix}
   -i & +1 \\
   +1 & +1
   \end{bmatrix}
$$

 $$ V^\dag V =
   \frac{1}{\sqrt{2}}\frac{1}{\sqrt{2}}
   \begin{bmatrix}
   -i & +1 \\
   +1 & +1
   \end{bmatrix}
   \begin{bmatrix}
   +i & -i \\
   +1 & +1
   \end{bmatrix}
   = \frac{1}{2}
   \begin{bmatrix}
   2 & 0 \\
   0 & 2
   \end{bmatrix}
   =
  \begin{bmatrix}
   1 & 0 \\
   0 & 1
   \end{bmatrix}
$$

As we see, when we multiply V with it's conjugate transpose, we get the identity matrix and thus V is unitary.
